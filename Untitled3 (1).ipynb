```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('colombianosexterior.csv', error_bad_lines=False)
```

```python
print(df.to_string())
```

```python
print(df.head(5000).to_string())
```

```python
import seaborn as sns
```

```python
df['Grupo edad'].value_counts().plot(kind='bar')
```

```python
print("Podemos ver que la gran mayoría de las personas que salen del país son adultos.")
```

```python
df['País'].value_counts().plot(kind='bar')
```

```python
print("Podemos observar que la mayoría de estos registros se encuentran en Argentina.")
```

```python
ax = sns.boxplot(x='Estado civil', y='Edad (años)', data=df)
```

```python
print("Como se muestra en la tabla, cerca de los 30 a 40 años la mayoría son solteros, cerca de los 40 a 50 años son casados, cerca de los 50 a 60 años hay mayor concentración de divorciados, y cerca de los 60 a 75 años la mayoría son separados.")
```

```python
sns.violinplot(x='Nivel Académico', y='Estatura (CM)', data=df)
plt.xticks(rotation=90)
```

```python
print("Podemos ver que no hay mucha disperción de los datos en cuanto al nivel académico, aunque se puede ver ligeramente mayor el nivel de bachillerato")
```

```python
sns.scatterplot(x='Área Conocimiento', y='Sub Area Conocimiento', hue='Género', data=df)
yticklabels = plt.gca().get_yticklabels()
plt.gca().set_yticks(range(0, len(yticklabels), 2))
plt.gca().set_yticklabels(yticklabels[::2], fontsize=8)
plt.xticks(rotation=90)
plt.show()
```

```python
Print("Podemos ver que la mayoría estudian entre ciencias sociales y humanas, donde la mayoría son mujeres, ciencias de la salud, también con mayoría en mujeres, e ingeniería arquitectura y afines, donde la mayoría son hombres.")
```
